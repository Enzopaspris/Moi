```python
import datetime
AJD=datetime.date.today() #Pour avoir la date d'aujourd'hui
An_actuel = AJD.year
Mois_actuel=AJD.month
Jour_actuel=AJD.day

class naissance:
    
    """ Création de la classe naissance 
    Avec j le jour, m le mois et a l'année qui sont tout le temps des entiers 
    
        Méthodes utilisées :
    
                    * anniversaire(self, j, m, a) --> None
                    * bissextile(self,j, m, a) --> None
                    * age(self, j, m, a)) --> None
                    * infos(self, j, m, a) --> None
            
         Attributs utilisés :
     
                    * self.j
                    * self.m
                    * self.a
        Test :
    
            >>> j = 25
            >>> m = 12
            >>> a = 2004
            >>> d = naissance(j, m, a)
            >>> d.age(j, m, a)
            'Tu as 16 ans.'
                
        """
               
    def __init__(self, j, m, a):
        """Méthodes permettant de construire la date"""
        
        self.a = a
        self.m = m
        self.j = j
    
    def anniversaire(self, j, m, a): 
        """Méthodes permettant de determiner dans combien de jours est l'anniversaire de l'utilisateur"""
        
        d0 = datetime.date(An_actuel ,self.m, self.j)
        d1 = datetime.date.today()
        delta = abs(d1 - d0)
        if d1 == d0:
            return f"Bon anniversaire !"
        else:
            return f"Ton anniversaire est dans {delta.days} jours (soit dans environ {delta.days//30} mois et {delta.days%30} jours)"

        
    def bissextile(self,j, m, a):
        """Méthodes permettant de savoir si oui ou non une année est bissextile""" 
        
        if int(self.a)>=1582:
            if (int(self.a)%4==0 and not int(self.a)%100==0) or int(self.a)%400==0:
                return f"{self.a} est une année bissextile "
            else :
                    return f"{self.a} n'est pas une année bissextile "
        else:
            return f"Votre année est inférieur a 1582"
        
    def age(self, j, m, a):
        """Méthodes permettant de determiner l'age de quelqu'un a partir de sa date de naissance"""
        
        if (self.m>Mois_actuel) or (self.m==Mois_actuel and self.j>Jour_actuel):
            Age= An_actuel-self.a-1
        else :
            Age = An_actuel-self.a
        return f"Tu as {Age} ans."
        
    
    def infos(self, j, m, a):
        """Renvoie 2 liens Wikipédia qui informe sur l'année et le mois de naissance de l'utilisateur"""
        
        if m>12:
            return f"Cette date n'existe pas" 
        if m == 1: #Transforme m en chaîne de caractères (pour le lien Wikipédia)
            m = 'Janvier'
        if m == 2:
            m = 'Février'
        if m == 3:
            m = 'Mars'
        if m == 4:
            m = 'Avril'
        if m == 5:
            m = 'Mai'
        if m == 6:
            m = 'Juin'
        if m == 7:
            m = 'Juillet'
        if m == 8:
            m = 'Août'
        if m == 9:
            m = 'Septembre'
        if m == 10:
            m = 'Octobre'
        if m == 11:
            m = 'Novembre'
        if m == 12:
            m = 'Décembre'
        return (f"Informations sur votre année de naissance : https://fr.wikipedia.org/wiki/{a} Informations sur votre mois de naissance https://fr.wikipedia.org/wiki/{m}_{a} ")

    def __repr__(self):
        return str(self.j, self.m, self.a)
    
    def __str__(self):
        return str(self.j, self.m, self.a)
    
if __name__ == "__main__":
        import doctest
        
        doctest.testmod()
    
```
