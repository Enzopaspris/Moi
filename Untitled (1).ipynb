```python

class Fraction():
    
    def _pgcd(a, b):
        if a < b:
            a, b = b, a
        while b > 0:
            a, b = b, a % b
        return a
    
    def addition(self, a, b):
        if b == self._den: 
            self._num = a + self._num
        else :
            Fraction = (a * self._den) + (self._num * b)
            print("L'addition des fractions est :" + str(Fraction) + '/' + str(b * self._den))       
            
    def soustraction(self, a, b):
        if b == self._den: 
            self._num = a - self._num
        else :
            Fraction = (a * self._den) - (self._num * b)
            print("L'addition des fractions est :" + str(Fraction) + '/' + str(b * self._den)) 
        
    
    def multiplication(self, a , b):
        self._num = self._num * a
        self._den = self._den * b

    def _simplifie(self):
        d = Fraction._pgcd(abs(self.num), abs(self.den))
        self._num = self.num // d
        self._den = self.den // d
    
    def __init__(self, a, b):
        self.num = a
        self.den = b
        self._num = None
        self._den = None
        self._simplifie()
        
    def inverse(self):
        self.num, self.den = self.den, self.num
        self._num, self._den = self._den, self._num
        
    def __str__(self):
        return f"{self.num} / {self.den}"
    
    def __repr__(self):
        return f"{self.num} / {self.den} = {self._num} / {self._den}"
```
