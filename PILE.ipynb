```python
class Pile:
    
    def __init__(self):
        self.mapile = []
    
    def __len__(self):
        return len(self.mapile)
         
    def sommet(self):
        if len(self) > 0 :
            return self.mapile[0]
    
    def empiler(self,element):
        self.mapile = [element] + self.mapile
    
    def depiler(self):
        if len(self) > 0:
            return self.mapile.pop(0)
```

```python
class Maillon:
    def __init__(self, element, suivant=None):
        # assert suivant is None or isinstance(suivant, Maillon)
        
        self.element = element
        self.suivant = suivant
        
class Liste_chainee():
    
    def __init__(self):
        """Initialise une liste vide."""
        
        self.premier = None
        
    def __len__(self):
        """Renvoie le nombre d'éléments présents dans la liste."""
        
        actuel = self.premier
        compteur = 0
        
        while actuel is not None:
            compteur += 1
            actuel = actuel.suivant
        
        return compteur
    
    def ajoute_en_tete(self, element):
        """Insère élément en tête de liste en créant un nouveau maillon"""
        
        self.premier = Maillon(element, self.premier)
        
    def supprime_en_tete(self):
        """Supprime l'élément en tête et retourne ce dernier"""
        
        if self.premier is not None:
            poubelle = self.premier
            self.premier = self.premier.suivant
            poubelle_temp = poubelle.element
            del poubelle
            return poubelle_temp
        
    def ajoute_en_queue(self, contenu):
        """Ajoute le contenu sous forme d'un maillon à 
        la fin de la liste chaînée
        """
        
        if self.premier is None:
            self.ajoute_en_tete(contenu)
        else:
            actuel = self.premier
            while actuel.suivant:
                actuel = actuel.suivant
            actuel.suivant = Maillon(contenu, None)
    
    def ieme_element(self, i):
        """accès au contenu du i-ème élément"""
        actuel = self.premier
        
        for _ in range(i - 1):
            actuel = actuel.suivant
        
        return actuel.element
            
```

```python
class Pile:
    """avec la liste chainée"""
    
    def __init__(self):
        self._lc = Liste_chainee()
        
    def __len__(self):
        return len(self._lc)
    
    def sommet(self):
        return self._lc.ieme_element(1)
    
    def empiler(self, element):
        self._lc.ajoute_en_tete(element)
        
    def depiler(self):
        return self._lc.supprime_en_tete()
        
```

```python
p = Pile()
p.empiler(3)
p.empiler(5)
print(p.sommet())
p.empiler(8)
p.depiler()
print(p.depiler())
print(p.depiler())
```
