```python
class Fraction():
    
    def _pgcd(a, b):
        if a < b:
            a, b = b, a
        while b > 0:
            a, b = b, a % b
        return a
    
    def _addition(a, b, c, d):
        if b == d: 
            nv_num = a + c
        else :
            Fraction = (a * d) + (c * b)
            print("L'addition des fractions est :" + str(Fraction) + '/' + str(b * d))       
        return nv_num
    
    def _multiplication(a, b, c, d):
        Fraction ="La multiplication des fractions est:" + int(a * c) + '/' + int(b * d)
        return Fraction

    def _simplifie(self):
        d = Fraction._pgcd(abs(self.num), abs(self.den))
        self._num = self.num // d
        self._den = self.den // d
    
    def __init__(self, a, b):
        self.num = a
        self.den = b
        self._num = None
        self._den = None
        self._simplifie()
        
    def inverse(self):
        self.num, self.den = self.den, self.num
        self._num, self._den = self._den, self._num
        
    def __str__(self):
        return f"{self.num} / {self.den}"
    
    def __repr__(self):
        return f"{self.num} / {self.den} = {self._num} / {self._den}"
```
