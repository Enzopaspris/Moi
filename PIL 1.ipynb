```python
from PIL import Image, ImageDraw


mon_image = Image.new("RGB", (800, 600), (250, 0, 0))
mon_image.save("rouge.jpg")

couche_g = ImageDraw.Draw(mon_image)
couche_g.line([(0, 0), (800, 600)], 
              fill=(255, 255, 0), width=20)
mon_image.save("rouge_diag_jaune.jpg")

couche_g.ellipse([(100,0), (700, 600)],
                 fill="navy",
                 outline="blue",
                 width=10)
mon_image.save("image.jpg")
```

```python
mon_image.show()
```
