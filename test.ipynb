```python
class Fraction():
    
    """ Création de la classe nommée "Fraction" 
    Méthodes utilisées :
    
    Avec a un entier et b un entier non nul
    
            Publiques :
                * addition(a, b) --> None
                * soustraction(a, b) --> None
                * multiplication(a, b) --> None
                * division(a, b) --> None
                * inverse() --> None
            
            Privées ( marqués par un '_' au début ) :
                * _pgcd(a, b) --> None
                * _simplifie() --> None
            
     Attributs utilisés :
    
            Publics :
                * self.num
                * self.den 
            
            Privés :
                * self._num
                * self._den
                * self._signe
"""

    def _pgcd(a, b):
        """Determiner le plus grand diviseur commun entre 2 valeurs"""
        
        if a < b:
            a, b = b, a
        while b > 0:
            a, b = b, a % b
        return a
    
    def _signe(self, a, b):
        if a / b < 0:
            self._signe = -1
        elif a / b == 0:
            self._signe = 0
        else : 
            self._signe = 1 
    
    def addition(self, a, b):
        """Additioner deux fractions"""
        
        if b == self._den: 
            self._num = a + self._num
        else :
            Fraction = (a * self._den) + (self._num * b)
            print("L'addition des fractions est : " + str(Fraction) + '/' + str(b * self._den))       
            
    def soustraction(self, a, b):
        """Soustraire deux fractions"""
        
        if b == self._den: 
            self._num = a - self._num
        else :
            Fraction = (a * self._den) - (self._num * b)
            print("La soustraction des fractions est : " + str(Fraction) + '/' + str(b * self._den)) 
            
    def multiplication(self, a , b):
        """Multiplier deux fractions"""
        
        self._num *= self.num
        self._den *= self.den
        print("La multiplication des fractions est : " + str(self.num) + '/' + str(self._den)) 
    
    def division(self, a, b):
        """Diviser deux fractions"""
        
        self._num *= self.den
        self._den *= self.num
        print("La division des fractions est : " + str(self.num) + '/' + str(self._den)) 

    def _simplifie(self):
        d = Fraction._pgcd(abs(self.num), abs(self.den))
        self._num = self.num // d
        self._den = self.den // d   
    
    def __init__(self, a, b):
        """Fonction permettant de construire la fraction"""
    
        self.num = a
        self.den = b
        self._num = None
        self._den = None
        self._simplifie()
        
    def inverse(self):
        """Inverse le numerateur et le dénominateur d'une fraction"""
        
        self.num, self.den = self.den, self.num
        self._num, self._den = self._den, self._num
        
    def __str__(self):
        return f"{self.num} / {self.den}"
    
    def __repr__(self):
        return f"{self.num} / {self.den} = {self._num} / {self._den}"
```
