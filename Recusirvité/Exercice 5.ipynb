```python
import doctest
def recherche(T, e):
    """
    >>> recherche([1, 4, 9, 3.54], 3.54)
    True
    >>> recherche([1, 4, 9, 3.54], 8)
    False
    """
    
    if len(T) == 0:
        return False
    elif len(T) == 1: 
        return T[0] == e
    return T[0] == e or recherche(T[1:], e)

def recherche2(T, e):
    """
    >>> recherche2([1, 4, 9, 3.54], 3.54)
    True
    >>> recherche2([1, 4, 9, 3.54], 8)
    False
    """
    
    if len(T) == 0: 
        return False
    elif len(T) == 1:
        return e == T[0]
    
    indice_milieu = len(T) // 2
    return recherche2(T[:indice_milieu], e) or recherche2(T[indice_milieu:], e)
    
doctest.testmod()
```
