```python
import doctest
def recherche_dicho(T, e):
    """
    >>> recherche_dicho([1, 2, 3, 4], 3)
    True
    >>> recherche_dicho([1, 2, 3, 4], 2.5)
    False
    """
    
    if len(T) == 0:
        return False
    indice_milieu = len(T) // 2
    if T[indice_milieu] == e:
        return True
    elif T[indice_milieu] > e:
        return recherche_dicho(T[:indice_milieu], e)
    else: 
        return recherche_dicho(T[indice_milieu+1:], e)

doctest.testmod()
```
